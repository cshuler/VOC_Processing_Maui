```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from io import StringIO


# Try and access file path of public data 
url = 'https://github.com/cshuler/VOC_Processing_Maui/blob/main/MASTER_Maui_VOC_Sheet_PUBLIC.csv?raw=true'
df = pd.read_csv(url,index_col=0)
print(df.head(5))
# Only have tapwater samples 
tap_water_variations = ['Tap Water', 'tap', 'Tap', 'tap water','Tap Water ', ' Tap Water','Tap water']
filter_water_variations = ['tap filtered water','Tap Water Filtered', 'tap water filtered', ' Tap Water Filtered', 'Tap Water Filtered ','Tap Water filtered']
#Filter the dataframe
tapwaterdf = df[df['SampleType'].isin(tap_water_variations)]
print(tapwaterdf)
filtered_waterdf = df[df['SampleType'].isin(filter_water_variations)]
id_mapping = {
    '03-11-24-01-JI': '03-11-24-JI-01',
    '03-20-24-JI-01': '3-20-24-JI-01',
    '03-20-24-JI-12': '3-20-24-JI-12',
    '11-07-CKS-05': '11-07-CKS-5',
    '03-20-CDS-10': '3-20-24-CES-10',
    '9-29-JES-1':'9-26-JES-1',
    '11-07-CKS-2':'11-07–CKS-2',
    '11-07-CKS-3':'11-07–CKS-3',
    "01-19-KDP-06":"01-10-KDP-06"
}

# Update the Sample IDs in all_df
tapwaterdf['Sample ID'] = tapwaterdf['Sample ID'].replace(id_mapping)
filtered_waterdf['Sample ID'] = filtered_waterdf['Sample ID'].replace(id_mapping)
def parse_datetime(date_string):
    # List of potential datetime formats
    formats = [
        "%m/%d/%Y %H:%M:%S",   # e.g., 9/8/2023 15:10:00
        "%m/%d/%Y %H:%M:%S %p",# e.g., 1/4/2024 5:02:00 PM
        "%m/%d/%Y",            # e.g., 8/30/2024
        "%m-%d-%y",            # e.g., 8-14-24
        "%m-%d-%Y %H:%M:%S",   # e.g., 10-18-2024 8:30:00
        "%m-%d-%Y",            # e.g., 10-18-2024
    ]
    
    for fmt in formats:
        try:
            return pd.to_datetime(date_string, format=fmt)
        except ValueError:
            continue
    # Return NaT if no format matches
    return pd.NaT
filtered_waterdf['DateTime Taken'] = filtered_waterdf['DateTime Taken'].apply(parse_datetime)
tapwaterdf['DateTime Taken'] = tapwaterdf['DateTime Taken'].apply(parse_datetime)

# ----------------------------------------------------------------------------------------------------#
# - Bring in clean VOC data -
# ----------------------------------------------------------------------------------------------------#

# Function to get the list of CSV files from the GitHub repository page
def get_csv_files_from_repo(api_url):
    response = requests.get(api_url)
    if response.status_code == 200:
        files = response.json()
        csv_files = [file['download_url'] for file in files if file['name'].endswith('.csv')]
        return csv_files
    else:
        raise ValueError(f"Failed to fetch repository contents: {response.status_code}")

# URL of the GitHub repository API
api_url = "https://api.github.com/repos/cshuler/VOC_Processing_Maui/contents/Clean_Data"

# Get the list of CSV files
csv_files = get_csv_files_from_repo(api_url)

# Check if the list of CSV files is populated
if not csv_files:
    raise ValueError("No CSV files found in the repository")

# Function to read a CSV file from a URL into a DataFrame
def read_csv_from_url(url):
    response = requests.get(url)
    response.raise_for_status()  # Raise an error for bad status codes
    return pd.read_csv(StringIO(response.text))

# Read all CSV files into a list of DataFrames and set 'Peak' as the index
dataframes = []
for url in csv_files:
    try:
        df = read_csv_from_url(url)
        df.set_index('Peak', inplace=True)
        dataframes.append(df)
    except Exception as e:
        print(f"Error reading {url}: {e}")
# Ensure there are DataFrames to concatenate
if not dataframes:
    raise ValueError("No DataFrames were read successfully")
# ----------------------------------------------------------------------------------------------------#
# - Calc detects -
# ----------------------------------------------------------------------------------------------------#

def clean_numeric_data(df):
    """Clean data by removing problematic characters and coercing to numeric."""
    # Remove any leading/trailing spaces
    df = df.applymap(lambda x: str(x).strip() if isinstance(x, str) else x)
    
    # Convert all values to numeric (float), coercing errors to NaN
    df = df.apply(pd.to_numeric)
    
    return df
def compare_samples_to_limit(df):
    # Identify the reporting limit column
    limit_col = [col for col in df.columns if 'Reporting_Limit' in col][0]
    # Identify the sample columns (all columns except the reporting limit)
    sample_cols = [col for col in df.columns if col != limit_col]
    
    # Compare and set sample values
    for sample_col in sample_cols:
        df[sample_col] = df.apply(lambda row: np.nan if row[sample_col] < row[limit_col] else row[sample_col], axis=1)
    
    return df

# Process each DataFrame
for i, df in enumerate(dataframes):
    df.replace(['n.a.', 'n.a./n.r.'], np.nan, inplace=True)
    df.index = df.index.to_series().replace('Methylene chloride', 'Methylene chloride (DCM)')
    df.index = df.index.str.strip()
    df = clean_numeric_data(df)
    dataframes[i] = compare_samples_to_limit(df)  
    
# Combine all DataFrames into one using an outer join to align the chemicals
combined_df = pd.concat(dataframes, axis=1, join='outer')
def TTHM_Calculator(row):
    TTHM_List = ['Bromodichloromethane', 'Bromoform', 'Dibromochloromethane', 'Trichloromethane (chloroform)']
    empty_List = []
    
    for chemical in TTHM_List:
        value = row.get(chemical, np.nan)
        empty_List.append(value)
    
    TTHM = np.nansum(empty_List)  # Sum the list using nansum to treat NaNs as 0s
    return TTHM
# Display the combined DataFrame
print("Combined DataFrame:")
print(combined_df.head())
```

```python
#Example for TTHMs

analysis_df.set_index('Peak', inplace=True)
chemical = "Carbon disulfide"
data = analysis_df.loc[chemical]
tap_samples = {col: val for col, val in data.items() if "(Tap)" in col}
filtered_samples = {col: val for col, val in data.items() if "(Filtered)" in col}
print(tap_samples)
#Match taps with corresponding filters based on order
comparisons = []
columns = list(data.index)  # Maintain column order

for i, tap_col in enumerate(columns):
    if "(Tap)" in tap_col:
        tap_val = data[tap_col]

        # Find the next filtered sample(s)
        matched_filters = []
        for j in range(i + 1, len(columns)):
            if "(Tap)" in columns[j]:  # Skip if another tap is directly after
                break
            if "(Filtered)" in columns[j]:  # Collect filtered matches
                matched_filters.append(columns[j])

        # Store comparisons
        for filt_col in matched_filters:
            comparisons.append((tap_col, filt_col, tap_val, data[filt_col]))

# Convert to DataFrame
comp_df = pd.DataFrame(comparisons, columns=["Tap Sample", "Filtered Sample", "Tap Value", "Filtered Value"])
comp_df[["Tap Value", "Filtered Value"]] = comp_df[["Tap Value", "Filtered Value"]].apply(pd.to_numeric)
comp_df['Difference']= comp_df['Tap Value'] - comp_df['Filtered Value']
comp_df['Percent Reduction'] = comp_df['Difference']/ comp_df['Tap Value'] *100
filtration_row = analysis_df.loc["Filtration"]

# Map the filtration type to the corresponding filtered sample in comp_df
comp_df["Filtration Type"] = comp_df["Filtered Sample"].map(filtration_row).fillna("Unknown")
comp_df.fillna(0, inplace=True)
comp_df
```

```python
import pandas as pd

# Assuming all_filtered_data is already defined
all_filtered_data.set_index('Peak', inplace=True)

# Chemicals to analyze
chemicals = ["Total Trihalomethanes", "Tetrahydrofuran"]

# Store all comparisons
all_comparisons = []

for chemical in chemicals:
    if chemical in all_filtered_data.index:
        data = all_filtered_data.loc[chemical]
        columns = list(data.index)  # Maintain column order

        comparisons = []
        pending_taps = []  # Store consecutive taps

        # Iterate through columns while maintaining order
        for i, col in enumerate(columns):
            if "(Tap)" in col:
                pending_taps.append(col)  # Store taps until a filtered sample appears
            elif "(Filtered)" in col:
                if pending_taps:  # Ensure there are taps before matching
                    # Match all stored taps to this filtered sample
                    for tap_col in pending_taps:
                        comparisons.append((chemical, tap_col, col, data[tap_col], data[col]))

                    # Continue checking for additional filtered samples (preserve original behavior)
                    last_tap = pending_taps[-1]  # Keep last tap for additional matches

                    for j in range(i + 1, len(columns)):
                        if "(Filtered)" in columns[j]:
                            comparisons.append((chemical, last_tap, columns[j], data[last_tap], data[columns[j]]))
                        elif "(Tap)" in columns[j]:  # Stop when a new tap appears
                            break

                    # Reset tap buffer after processing all available filters
                    pending_taps = []

        # Append results for this chemical
        all_comparisons.extend(comparisons)

# Convert to DataFrame
comp_df_2 = pd.DataFrame(all_comparisons, columns=["Chemical", "Tap Sample", "Filtered Sample", "Tap Value", "Filtered Value"])
comp_df_2[["Tap Value", "Filtered Value"]] = comp_df_2[["Tap Value", "Filtered Value"]].apply(pd.to_numeric)
comp_df_2['Difference'] = comp_df_2['Tap Value'] - comp_df_2['Filtered Value']
comp_df_2['Percent Reduction'] = (comp_df_2['Difference'] / comp_df_2['Tap Value']) * 100

# Fill NaN values (if any) with 0
comp_df_2.fillna(0, inplace=True)
comp_df_2
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define colors for each chemical
chemical_colors = {"Total Trihalomethanes": "purple", "Tetrahydrofuran": "orange"}

# Generate x-axis labels as "Pair 1, Pair 2, ..."
tap_samples = comp_df_2["Tap Sample"].unique()
pair_labels = [f"Pair {i+1}" for i in range(len(tap_samples))]

# X positions for bars
x = np.arange(len(pair_labels))

# Aggregate by tap sample and chemical
agg_df = comp_df_2.groupby(["Tap Sample", "Chemical"])["Difference"].sum().unstack(fill_value=0)

# Ensure all chemicals exist in the DataFrame
for chem in chemical_colors.keys():
    if chem not in agg_df.columns:
        agg_df[chem] = 0  # Add missing chemical column with zero values

# Reorder according to tap sample appearance
agg_df = agg_df.reindex(tap_samples)

# Create stacked bar plot
fig, ax = plt.subplots(figsize=(12, 6))

bottom = np.zeros(len(agg_df))  # Initialize bottom position for stacking

for chem in chemical_colors.keys():
    ax.bar(x, agg_df[chem], color=chemical_colors[chem], label=chem, bottom=bottom)
    bottom += agg_df[chem]  # Update bottom position for stacking

# Labels and title
ax.set_xticks(x)
ax.set_xticklabels(pair_labels, rotation=90, fontsize=8)
ax.set_ylabel("Difference (Tap - Filter)")
ax.set_title("Stacked Difference from Tap - Filter")

# Add legend
ax.legend(title="Chemical")

plt.tight_layout()
plt.show()
```
